```python
def Initial_condition(den):
    Graph=nx.read_gpickle('Graph.gpickle')
    for i in Graph.nodes:
        if den>np.random.rand():
            Graph.nodes[i]['x']=1
        else:
            Graph.nodes[i]['x']=0
    x=np.random.randint(50)
    y=np.random.randint(50)
    Graph.nodes[x,y]['x']=2
    return Graph
```

```python
def Move_it():
    for i in G.nodes:
        if G.nodes[i]['x']!=0:
            ret=0
            adjnum=4
            adj=list(G.adj[i])
            while ret==0 and adjnum>0:
                r=np.random.randint(adjnum)
                if adj[r]==0:
                    G.nodes[adj[r]]['x']=G.nodes[i]['x']
                    G.nodes[i]['x']=0
                    ret=1
                adj.remove(adj[r])
                adjnum-=1
```

```python
def Count_nodes(Graph):
    snode=[]
    inode=[]
    rnode=[]
    for n_1 in Graph.nodes:
        if Graph.nodes[n_1]['x']==1:
            snode.append(n_1)
        if Graph.nodes[n_1]['x']==2:
            inode.append(n_1)  
        if Graph.nodes[n_1]['x']==3:
            rnode.append(n_1)
    return (snode,inode,rnode)
```

```python
def Transmition(d,p,q,snode,inode):
    for i in snode:
        for j in inode:
            if nx.shortest_path_length(G, source=i, target=j)<d and p>np.random.rand():
                G.nodes[i]['x']=2
    for i in inode:
        if q>np.random.rand():
            G.nodes[i]['x']=3
```

```python
def dynamic(d,p,q):
    (snode,inode,rnode)=Count_nodes(G)
    while len(inode)!=0:
        Move_it()
        (snode,inode,rnode)=Count_nodes(G)
        Transmition(d,p,q,snode,inode)
    return len(rnode)
```

```python
plist=[0.2,0.4,0.6,0.8,1]
rdata=[[],[],[],[],[]]
for p in plist:
    for i in range(1000):
        G=Initial_condition(0.02)
        rdata[plist.index(p)].append(dynamic(15,p,1)) 
```

```python
Cs = []
for r in rdata:
    c = []
    for n in set(r):
        c.append(r.count(n))
    Cs.append(c)    
```

```python
plt.figure(figsize=(10,7),dpi=100)
plt.grid(color = 'g', linestyle = '--', linewidth = 0.5)
plt.title('D=15')
plt.xlabel('Transition probability')
plt.ylabel('Final Recovered number')
for i in range(5):
    plt.scatter(plist[i]*np.ones_like(Cs[i]),list(set(rdata[i])),s=Cs[i],marker='o')
plt.legend(loc=0)
plt.savefig('fig1-1')
```
